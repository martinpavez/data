This notebook shows the a comparison between data from 10 selected stations from two different datasets. 

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
os.chdir("../")
print(os.getcwd())
```

## Code

### Monthly

```python
hws_cimis = pd.read_csv("data/local_data/monthly/HWs_california_CIMIS.csv")
hws_cimis["Date"] = pd.to_datetime(hws_cimis["Date"],format='%Y-%m-%d')
hws_cimis.set_index('Date', inplace=True)
hws_noaa = pd.read_csv("data/local_data/monthly/HWs_california_NOAA.csv")
hws_noaa["Date"] = pd.to_datetime(hws_noaa["Date"],format='%Y-%m')
hws_noaa.set_index('Date', inplace=True)
```

```python
indices = ["HWN", "HWF", "HWD", "HWM", "HWA"]
fix, axs = plt.subplots(5, 1, figsize=(30,30))
for i in range(5):
    corr = np.corrcoef(hws_noaa[hws_noaa.index.year >= 1989][f"{indices[i]}"], hws_cimis[f"{indices[i]}"])[0,1]
    axs[i].plot(hws_noaa.index, hws_noaa[f"{indices[i]}"], label = "NOAA")
    axs[i].plot(hws_cimis.index, hws_cimis[f"{indices[i]}"], label = "CIMIS")
    axs[i].set_title(f"Index {indices[i]}, Correlation from 1989: {np.round(corr, 2)}")
    axs[i].legend()
```
